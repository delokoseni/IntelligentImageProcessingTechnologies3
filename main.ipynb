```python
def printTensorInfo(tensor, title):
    print(f"====={title}=====")
    print(f"tensor:\n{tensor}")
    print(f"shape: {tensor.shape}")
    print(f"dtype: {tensor.dtype}")
    print(f"type: {tensor.type()}")
    print(f"dim: {tensor.dim()}")
```

```python
# Различные способы создания тензоров
tensorRecommended = torch.Tensor(3, 5, 2)
printTensorInfo(tensorRecommended, "tensorRecommended")
tensorNotRecommended = torch.empty(3, 5, 2)
printTensorInfo(tensorNotRecommended, "tensorNotRecommended")
tensorZeros = torch.zeros(3, 5, 2)
printTensorInfo(tensorZeros, "tensorZeros")
tensorOnes = torch.ones(3, 5, 2)
printTensorInfo(tensorOnes, "tensorOnes")
tensorRand = torch.rand(3, 5, 2)
printTensorInfo(tensorRand, "tensorRand")
tensorSpecificDtype = torch.empty(3, 5, 2, dtype=torch.uint8)
printTensorInfo(tensorSpecificDtype, "tensorSpecificDtype")
tensorClass = torch.IntTensor(3, 5, 2)
printTensorInfo(tensorClass, "tensorClass")
```

```python
# Различные способы создания тензоров с разными значениями
firstTensor = torch.tensor([1])
printTensorInfo(firstTensor, "firstTensor")
secondTensor = torch.tensor([1, 2.0])
printTensorInfo(secondTensor, "secondTensor")
thirdTensor = torch.tensor([[1, 2], [3, 4], [5, 6]])
printTensorInfo(thirdTensor, "thirdTensor")
fourthTensor = torch.tensor([[1, 2], [3, 4], [5, 6]], dtype=torch.float32)
printTensorInfo(fourthTensor, "fourthTensor")
```

```python
# Еще способы заполнить тензор
tensorEye = torch.eye(3)
printTensorInfo(tensorEye, "tensorEye")
tensorFull = torch.full((2, 4), 5)
printTensorInfo(tensorFull, "tensorFull")
tensorArange = torch.arange(-5, 0, 2)
printTensorInfo(tensorArange, "tensorArange")
tensorFull.fill_(4) # Только inplace
printTensorInfo(tensorFull, "tensorFull")
```

```python
# Методы преобразования формы тензоров
exampleTensor = torch.arange(27)
printTensorInfo(exampleTensor, "exampleTensor")
view = exampleTensor.view((3, 9))
printTensorInfo(view, "view")
view2 = exampleTensor.view((3, 3, 3))
printTensorInfo(view2, "view2")
reshape = exampleTensor.reshape(3, 3, 3)
printTensorInfo(reshape, "reshape")

resize = exampleTensor.resize_((2, 3))
exampleTensor[0, 0] = 777
printTensorInfo(resize, "resize")

ravel = exampleTensor.ravel()
printTensorInfo(ravel, "ravel")
exampleTensorT = exampleTensor.permute(1, 0) # Меняем оси местами
printTensorInfo(exampleTensorT, "exampleTensorT")
exampleTensorT2 = exampleTensor.mT # Меняем оси местами
printTensorInfo(exampleTensorT2, "exampleTensorT2")

```

```python
# Разложение Холецкого
def printMatrixInfo(A, name):
    print(f"{name}:\n{A}")
    print(f"shape: {A.shape}")
    print()
```

```python
A = torch.tensor([[4.0, 2.0], # Симметричная положительно определённая матрица
                  [2.0, 3.0]])
printMatrixInfo(A, "A")
L = torch.linalg.cholesky(A) # Разложение Холецкого
printMatrixInfo(L, "L (нижнетреугольная)")
A_reconstructed = L @ L.T # Проверка: A = L @ L.T
printMatrixInfo(A_reconstructed, "L @ L.T")
```

```python
print("Проверка разложения:\n")
matrices = [
    torch.tensor([[4.0, 2.0], [2.0, 3.0]]),
    torch.tensor([[9.0, 3.0], [3.0, 5.0]])
]
for i, A in enumerate(matrices):
    L = torch.linalg.cholesky(A)
    A_rec = L @ L.T
    print(f"Матрица {i+1}:")
    print("A:")
    print(A)
    print("L @ L.T:")
    print(A_rec)
    diff = torch.max(torch.abs(A - A_rec))
    print(f"Максимальная ошибка: {diff.item()}")
```

```python
print("Попытка разложения плохой матрицы:\n")
bad_matrix = torch.tensor([[1.0, 2.0],
                           [2.0, 1.0]])  # не положительно определённая
try:
    L = torch.linalg.cholesky(bad_matrix)
except RuntimeError as e:
    print("Ошибка:")
    print(e)
```

```python
print("Создание корректной матрицы:\n")
X = torch.randn(3, 3) # Любая матрица
A = X @ X.T # Делаем её положительно определённой
printMatrixInfo(A, "A")
L = torch.linalg.cholesky(A)
printMatrixInfo(L, "L")
```

```python
print("Решение системы Ax = b:\n")
A = torch.tensor([[4.0, 2.0], [2.0, 3.0]])
b = torch.tensor([1.0, 1.0])
L = torch.linalg.cholesky(A)
y = torch.linalg.solve(L, b) # Решаем: Ly = b
x = torch.linalg.solve(L.T, y) # Решаем: L^T x = y
print("Решение x:")
print(x)
print("\nПроверка Ax:")
print(A @ x)
```

```python
import os
import json
from PIL import Image

import torch
import torch.utils.data as data
import torchvision.transforms.v2 as tfs
import torch.nn as nn
import torch.optim as optim
from tqdm import tqdm
```

```python
class AlphabetDataset(data.Dataset):
    def __init__(self, path, train=True, transform=None):
        self.path = os.path.join(path, "train" if train else "test")
        self.transform = transform

        with open(os.path.join(path, "format.json"), "r") as fp:
            self.format = json.load(fp)

        self.length = 0
        self.files = []
        self.targets = torch.eye(26)

        for _dir, _target in self.format.items():
            path = os.path.join(self.path, _dir)
            list_files = os.listdir(path)
            self.length += len(list_files)
            self.files.extend(map(lambda _x: (os.path.join(path, _x), _target), list_files))

    def __getitem__(self, index):
        path_file, target = self.files[index]
        t = self.targets[target]
        img = Image.open(path_file)
        if self.transform:
            img = self.transform(img).ravel().float() / 255.0
        return img, t

    def __len__(self):
        return self.length
```

```python
class AlphabetNN(nn.Module):
    def __init__(self, input_dim, num_hidden, output_dim):
        super().__init__()
        self.layer1 = nn.Linear(input_dim, num_hidden)
        self.layer2 = nn.Linear(num_hidden, output_dim)

    def forward(self, x):
        x = self.layer1(x)
        x = nn.functional.relu(x)
        x = self.layer2(x)
        return x
```

```python
model = AlphabetNN(28*28, 32, 26)
to_tensor = tfs.ToImage()
d_train = AlphabetDataset("emnist_letters_folder", transform=to_tensor)
train_data = data.DataLoader(d_train, batch_size=32, shuffle=True)

optimizer = optim.Adam(params=model.parameters(), lr=0.001)
loss_function = nn.CrossEntropyLoss()

epochs = 2
model.train()
```

```python
for _e in range(epochs):
    loss_mean = 0 # Среднее значение функции потерь по эпохе
    lm_count = 0 # Текущее количество слагаемых
    train_tqdm = tqdm(train_data, leave=True)
    for x_train, y_train in train_tqdm:
        predict = model(x_train)
        loss = loss_function(predict, y_train)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        lm_count += 1
        loss_mean = 1 / lm_count * loss.item() + (1 - 1 / lm_count) * loss_mean
        train_tqdm.set_description(f"Epoch [{_e + 1}/{epochs}], loss_mean={loss_mean:.3f}")

d_test = AlphabetDataset("emnist_letters_folder", train=False, transform=to_tensor)
test_data = data.DataLoader(d_test, batch_size=500, shuffle=False)
Q = 0

model.eval()

for x_test, y_test in test_data:
    with torch.no_grad():
        p = model(x_test)
        p = torch.argmax(p, dim=1)
        y = torch.argmax(y_test, dim=1)
        Q += torch.sum(p == y).item()

Q /= len(d_test)
print(Q)
```
